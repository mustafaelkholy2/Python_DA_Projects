# Data Cleaning

### Drop Row have a null Values

```python
df.drop(33,inplace=True)
```

### Fill the null valus in __year_resale_value column with 0

```python
df.loc[(df['__year_resale_value'].isnull()), ['__year_resale_value']] = 0
```

### Group By Manufacturer Column

```python
df_groupby_Manufacturer = df.groupby(by='Manufacturer')
```

### Fill all null Values

```python
columns = []
for i in df.columns:
    if df[i].isnull().values.any():
        columns.append(i)
columns_have_null_value = np.array(columns)
columns_have_null_value
```

```python
for i in columns_have_null_value:
    Manufacturer_has_null = np.array(df[df[i].isnull()]['Manufacturer'])
    for j in Manufacturer_has_null:
        df.loc[(df['Manufacturer'] == j) & (df[i].isnull()), [i]] = df_groupby_Manufacturer.mean(numeric_only= True).loc[j,i]
```

```python
df.info()
```

## Another Way

```python
Manufacturer_has_null = np.array(df[df['Power_perf_factor'].isnull()]['Manufacturer'])
Manufacturer_has_null
```

```python
for i in Manufacturer_has_null:
    df.loc[(df['Manufacturer'] == i) & (df.Power_perf_factor.isnull()), ['Power_perf_factor']] = df_groupby_Manufacturer.mean(numeric_only= True).loc[i,'Power_perf_factor']
```

```python
df[(df['Manufacturer'].isin(Manufacturer_has_null))]
```

### Fill the null valus in Fuel_efficiency column with the average of Fuel_efficiency from the same Manufacturer

```python
Manufacturer_has_null = np.array(df[df['Fuel_efficiency'].isnull()]['Manufacturer'])
Manufacturer_has_null
```

```python
for i in Manufacturer_has_null:
    df.loc[(df['Manufacturer'] == i) & (df.Power_perf_factor.isnull()), ['Fuel_efficiency']] = df_groupby_Manufacturer.mean(numeric_only= True).loc[i,'Fuel_efficiency']
```

```python
df[(df['Manufacturer'].isin(Manufacturer_has_null))]
```
